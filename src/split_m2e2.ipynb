```python
import json
from random import random
```

```python
input_manifest = '/data/raw/m2e2_image_url_caption.json'
with open(input_manifest) as f:
    id2file = {}
    original_manifest = json.load(f)
```

```python
total_len = 0
valid_dict = {}
for article in original_manifest:
    keys_to_pop = []
    for img in original_manifest[article]:
        if random() > 0.8:
            if not article in valid_dict:
                valid_dict[article]={}
            valid_dict[article][img]=original_manifest[article][img]
            keys_to_pop.append(img)
    for keys in keys_to_pop:
        del original_manifest[article][keys]
    total_len+=len(original_manifest[article])
```

```python
with open('/data/raw/m2e2_train_caption.json', 'w') as f:
    json.dump(original_manifest, f)
    
with open('/data/raw/m2e2_valid_caption.json', 'w') as f:
    json.dump(valid_dict, f)
```
